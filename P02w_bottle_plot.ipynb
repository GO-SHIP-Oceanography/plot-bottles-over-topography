```python
%matplotlib inline
import numpy as np
import pandas as pd
import xarray as xr
import matplotlib.pyplot as plt
import gsw

from scipy.interpolate import interp1d
```

```python
class util:
    """some utility functions"""
    
    
    
    def track_dis(trk_lon, trk_lat):
        import gsw
        trk_dis = np.zeros(len(trk_lon), )
        for i in np.arange(1, len(trk_lon)):
            trk_dis[i] = gsw.distance([trk_lon[i], trk_lon[i-1]], [trk_lat[i], trk_lat[i-1]], p=0, axis=0)

        return trk_dis

    
    
    def track_dis_lon(trk_lon, trk_lat):
        import gsw
        trk_dis = np.zeros(len(trk_lon), )
        for i in np.arange(1, len(trk_lon)):
            trk_dis[i] = gsw.distance([trk_lon[i], trk_lon[i-1]], [trk_lat[i-1], trk_lat[i-1]], p=0, axis=0)

        return trk_dis

    
    
    def track_dis_lat(trk_lon, trk_lat):
        import gsw
        trk_dis = np.zeros(len(trk_lon), )
        for i in np.arange(1, len(trk_lon)):
            trk_dis[i] = gsw.distance([trk_lon[i-1], trk_lon[i-1]], [trk_lat[i], trk_lat[i-1]], p=0, axis=0)

        return trk_dis

    
    
    def nan_index_in_var(var):
        """ find leading and trailing indexes for nans in a 1-d array (var) """
        if len(var.shape)>1:
            raise ValueError('var must be a 1-d array')
        else:
            mask = np.isnan(var)
            leading_nans = mask.argmin()
            trailing_nans = mask[::-1].argmin()

            leading_nans_idx = np.arange(leading_nans)
            trailing_nans_idx = np.arange(var.size - trailing_nans, var.size)

            mask[leading_nans_idx] = 0
            mask[trailing_nans_idx] = 0
            middle_nans_idx = np.where(mask==1)[0]

        return leading_nans_idx, trailing_nans_idx, middle_nans_idx
    
    
    
    def m_per_nm():
        return 1851.85
    
    
    
    def fine_trk(lon_trk, lat_trk, n):
        lon_trk_new = []
        lat_trk_new = []
        for i in range(len(lon_trk)-1):
            lon_trk_ = np.linspace(lon_trk[i], lon_trk[i+1], n)[0:-1]
            lat_trk_ = np.linspace(lat_trk[i], lat_trk[i+1], n)[0:-1]
            lon_trk_new = np.append(lon_trk_new, lon_trk_)
            lat_trk_new = np.append(lat_trk_new, lat_trk_)
        return lon_trk_new, lat_trk_new

    
    
    def depth_trk(ds, lon_trk, lat_trk):
        depth = np.zeros(lon_trk.shape) + np.nan
        for i in range(len(depth)):
            depth[i] = ds.z.interp(lat=lat_trk[i], lon=lon_trk[i], method='linear')    
        return depth
    
    
    
    def lon_per_nm(lon_ref, lat_ref):
        return 1/(util.track_dis([lon_ref-.5, lon_ref+.5], [lat_ref, lat_ref])[1]/util.m_per_nm())
```

```python
def trk_info(trk, mask):
    station = trk["station"].to_numpy()[mask]
    lon = trk["longitude"].to_numpy()[mask]
    lon[lon<0] += 360
    lat = trk["latitude"].to_numpy()[mask]
    depth = trk["depth"].to_numpy()[mask]
    distance = util.track_dis(lon, lat)
    distance_nm = distance/util.m_per_nm()
    
    return {'station': station, 
            'longitude': lon, 
            'latitude': lat, 
            'depth': depth, 
            'distance': distance, 
            'distance_nm': distance_nm}
```

```python
def topo_along_section(ds_P02, P02_occupation, n=10):
    """get fine lon, lat, and distance
    """

    lon_fine, lat_fine = eval(f"util.fine_trk(P02_occupation['longitude'], P02_occupation['latitude'], {n})")
    depth_fine = util.depth_trk(ds_P02, lon_fine, lat_fine) 
    dis_fine = util.track_dis(lon_fine, lat_fine)
    dis_fine_nm = dis_fine/util.m_per_nm()
    
    return lon_fine, lat_fine, dis_fine_nm, depth_fine



def plot_bottle(ax, ds_P02, P02_occupation, lon_fine, lat_fine, depth_fine, dis_fine_nm, param_figure, param_symbols, method_dis = 0, method_with_real=1, method_text=1):
    """make figure for stations on top of bathymetry, with bottle depths on it
    
       method_dis = 1 [x axis is distance from first station], 0 [x axis is lon] [default]
       method_with_real = 1 [past stations plot real and design bottle depths] [default], 0 [off]
       method_text = 1 [station numbers on] [default], 0 [off]
    """
    from scipy.interpolate import interp1d
    
    if method_dis:
        # plot bottom depth along section
        ax.plot(np.cumsum(dis_fine_nm), depth_fine, 'k')
        ax.set_xlabel('Distance [nm]')
        f = interp1d(lon_fine, np.cumsum(dis_fine_nm), bounds_error=False, fill_value=np.nan)
        if np.isnan(f(param_figure['lon_range'][0])):
            ax.set_xlim([0, f(param_figure['lon_range'][1])]) 
        else:
            ax.set_xlim([f(param_figure['lon_range'][0]), f(param_figure['lon_range'][1])]) 
        # plot stations    
        mask = eval("(P02_occupation['longitude']>param_figure['lon_range'][0]) & (P02_occupation['longitude']<param_figure['lon_range'][1])")
        for i in range(len(eval("P02_occupation['longitude'][mask]"))):
            ax.plot([np.cumsum(P02_occupation['distance_nm'])[mask][i], np.cumsum(P02_occupation['distance_nm'])[mask][i]], [0, -P02_occupation['depth'][mask][i]], c='k', lw=.5)
            if method_text:
                ax.text(np.cumsum(P02_occupation['distance_nm'])[mask][i], param_figure['depth_range'][1]+(param_figure['depth_range'][0]-param_figure['depth_range'][1])/10, P02_occupation['station'][i])
        # plot bottles
        if method_with_real:
            for i in range(len(GOSHIP_blt['station'])):
                if GOSHIP_blt['status'][i][0]:
                    for j in range(len(GOSHIP_blt['bottle_number'][i][0])):
                        ax.plot(f(GOSHIP_blt['lon'][i][0]), -GOSHIP_blt['bottle_depth_real'][i][0][j], 
                                param_symbols['past_real'][0], 
                                c = param_symbols['past_real'][1],
                                mec = param_symbols['past_real'][2],
                                ms = param_symbols['past_real'][3],
                                mew = param_symbols['past_real'][-1],)  
                else:
                    for j in range(len(GOSHIP_blt['bottle_number'][i][0])):
                        ax.plot(f(GOSHIP_blt['lon'][i][0]), -GOSHIP_blt['bottle_depth_design'][i][0][j], 
                                param_symbols['todo_design'][0], 
                                c = param_symbols['todo_design'][1],
                                mec = param_symbols['todo_design'][2],
                                ms = param_symbols['todo_design'][3],
                                mew = param_symbols['todo_design'][-1],)  
        else:    
            for i in range(len(GOSHIP_blt['station'])):
                if GOSHIP_blt['status'][i][0]:
                    for j in range(len(GOSHIP_blt['bottle_number'][i][0])):
                        ax.plot(f(GOSHIP_blt['lon'][i][0]), -GOSHIP_blt['bottle_depth_design'][i][0][j], 
                                param_symbols['past_design'][0], 
                                c = param_symbols['past_design'][1],
                                mec = param_symbols['past_design'][2],
                                ms = param_symbols['past_design'][3],
                                mew = param_symbols['past_design'][-1],
                                alpha=1)
                else:
                    for j in range(len(GOSHIP_blt['bottle_number'][i][0])):
                        ax.plot(f(GOSHIP_blt['lon'][i][0]), -GOSHIP_blt['bottle_depth_design'][i][0][j], 
                                param_symbols['todo_design'][0], 
                                c = param_symbols['todo_design'][1],
                                mec = param_symbols['todo_design'][2],
                                ms = param_symbols['todo_design'][3],
                                mew = param_symbols['todo_design'][-1],)      
    else:
        # plot bottom depth along section
        ax.plot(lon_fine, depth_fine, 'k')
        ax.set_xlabel('Longitude [$^{\circ}$E]')
        ax.set_xlim([param_figure['lon_range'][0], param_figure['lon_range'][1]]) 
        # plot stations    
        mask = eval("(P02_occupation['longitude']>param_figure['lon_range'][0]) & (P02_occupation['longitude']<param_figure['lon_range'][1])")
        for i in range(len(eval("P02_occupation['longitude'][mask]"))):
            ax.plot([P02_occupation['longitude'][mask][i], P02_occupation['longitude'][mask][i]], [0, -P02_occupation['depth'][mask][i]], c='k', lw=.5)
            if method_text:
                ax.text(P02_occupation['longitude'][mask][i], param_figure['depth_range'][1]+(param_figure['depth_range'][0]-param_figure['depth_range'][1])/10, P02_occupation['station'][i])
        # plot bottles
        if method_with_real:
            for i in range(len(GOSHIP_blt['station'])):
                if GOSHIP_blt['status'][i][0]:
                    for j in range(len(GOSHIP_blt['bottle_number'][i][0])):
                        ax.plot(GOSHIP_blt['lon'][i][0], -GOSHIP_blt['bottle_depth_real'][i][0][j], 
                                param_symbols['past_real'][0], 
                                c = param_symbols['past_real'][1],
                                mec = param_symbols['past_real'][2],
                                ms = param_symbols['past_real'][3],
                                mew = param_symbols['past_real'][-1],)
                else:
                    for j in range(len(GOSHIP_blt['bottle_number'][i][0])):
                        ax.plot(GOSHIP_blt['lon'][i][0], -GOSHIP_blt['bottle_depth_design'][i][0][j], 
                                param_symbols['todo_design'][0], 
                                c = param_symbols['todo_design'][1],
                                mec = param_symbols['todo_design'][2],
                                ms = param_symbols['todo_design'][3],
                                mew = param_symbols['todo_design'][-1],)  
        else:    
            for i in range(len(GOSHIP_blt['station'])):
                if GOSHIP_blt['status'][i][0]:
                    for j in range(len(GOSHIP_blt['bottle_number'][i][0])):
                        ax.plot(GOSHIP_blt['lon'][i][0], -GOSHIP_blt['bottle_depth_design'][i][0][j], 
                                param_symbols['past_design'][0], 
                                c = param_symbols['past_design'][1],
                                mec = param_symbols['past_design'][2],
                                ms = param_symbols['past_design'][3],
                                mew = param_symbols['past_design'][-1],
                                alpha=1)
                else:
                    for j in range(len(GOSHIP_blt['bottle_number'][i][0])):
                        ax.plot(GOSHIP_blt['lon'][i][0], -GOSHIP_blt['bottle_depth_design'][i][0][j], 
                                param_symbols['todo_design'][0], 
                                c = param_symbols['todo_design'][1],
                                mec = param_symbols['todo_design'][2],
                                ms = param_symbols['todo_design'][3],
                                mew = param_symbols['todo_design'][-1],)      
                
    ax.set_ylim([param_figure['depth_range'][0], param_figure['depth_range'][1]])
    ax.grid(True)
    ax.set_ylabel('z [m]')
    ax.set_title(f'Now at {np.datetime64("now").astype("datetime64[m]")} UTC')
    
```
